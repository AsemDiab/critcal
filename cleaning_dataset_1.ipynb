```python
df = df.replace("اوافق", "4")
df = df.replace("أوافق بشدة", "5")
df = df.replace("محايد", "3")
df = df.replace("اعارض", "2")
df = df.replace("اعارض بشدة", "1")
df = df.replace("لا", "0")
df = df.replace("نعم", "1")

df
```

```python
arabic_to_western = {
    '٠': '0', '١': '1', '٢': '2', '٣': '3', '٤': '4',
    '٥': '5', '٦': '6', '٧': '7', '٨': '8', '٩': '9'
}

# Function to replace Arabic-Indic numerals with Western numerals
def convert_arabic_to_western(text):
    # Ensure the input is a string
    text = str(text)
    # Replace each Arabic-Indic digit with the corresponding Western digit
    for arabic_digit, western_digit in arabic_to_western.items():
        
        text = text.replace(arabic_digit, western_digit)
        if (text=="40 سنه"):
            text="40"
    return text
df['العمر'] = df['العمر'].apply(convert_arabic_to_western)

# Convert to integer if necessary and check if the number is in the range [0, 100]
df['العمر'] = df['العمر'].astype(float)
```

```python
df=df.replace(3.5,35)
```
