```python
def draw_pie(df,the_column_name):
    values=df[the_column_name].unique()
    counts=[]
    for value in values:
        counts.append(df[df[the_column_name]==value].shape[0])
    percent=[]
    for count in counts:
        percent.append(math.floor(count*10000/sum(counts))/100) 
    percent[-1]=math.floor(10000-100*sum(percent[0:-1]))/100
    chart_labels=[]
    for i in range(len(values)):
        print(values[i])
        print(translation_dict.get(values[i]))
        chart_labels.append(translation_dict.get(values[i])+" "+str(percent[i])+" %")    
    return plt.pie(counts,labels=chart_labels)
    
```
