```python
import pandas as pd
from scipy.stats import spearmanr
from scipy.stats import chi2_contingency
from scipy.stats import mannwhitneyu
the_result=[]

all_values = {
    'column': [],
    'P-value': [],
    'Mann-Whitney U statistic': [],
    'Chi-square': [],
    "Spearman Correlation Coefficient":[],
    "relation":[]
}

all_values_df = pd.DataFrame(all_values)

# Save DataFrame to CSV
```

```python
def Spearman_Rank_Correlation(x,y,col,ReturnA=True):
    # Calculate Spearman Rank Correlation
    correlation, p_value = spearmanr(x, y)

    print(f"Spearman Correlation Coefficient: {correlation}")
    print(f"P-value: {p_value}")

    # Interpretation
    if p_value < 0.05:
        print("The correlation is statistically significant.")
    else:
        print("The correlation is not statistically significant.")
    new_row = {
    'column': col,
    'P-value': p_value,
    'Mann-Whitney U statistic': "-",
    'Chi-square':"-",
    "Spearman Correlation Coefficient": correlation,
    "relation": 'Significant' if p_value < 0.05 else 'not Significant'
}

# Convert new row to DataFrame
    

    if ReturnA:
        return p_value < 0.05,new_row
```

```python
def Chi_Square_test(x,y,col,return_answer=True):
    contingency_table = pd.crosstab(x,y)

    # Perform Chi-square test
    chi2, p_value, dof, expected = chi2_contingency(contingency_table)

    print(f"Chi-square statistic: {chi2}")
    print(f"P-value: {p_value}")
    print(f"Degrees of freedom: {dof}")
    print("Expected frequencies:\n", expected)
    
    new_row = {
    'column': col,
    'P-value': p_value,
    'Mann-Whitney U statistic': "-",
    'Chi-square':chi2,
    "Spearman Correlation Coefficient": "-",
    "relation": 'Significant' if p_value < 0.05 else 'not Significant'
}


# Interpretation
    if p_value < 0.05:
        print("There is a significant relationship between City and Acceptance level.")
    else:
        print("There is no significant relationship between City and Acceptance level.")
    if return_answer:
        return p_value < 0.05,new_row
```

```python
def Mann_Whitney_U_Test(x,y,col,return_answer=True):
   

    # Perform Mann-Whitney U Test
    u_statistic, p_value = mannwhitneyu(x, y)

    print(f"Mann-Whitney U statistic: {u_statistic}")
    print(f"P-value: {p_value}")

    
    new_row = {
    'column': col,
    'P-value': p_value,
    'Mann-Whitney U statistic': u_statistic,
    'Chi-square':"-",
    "Spearman Correlation Coefficient": "-",
    "relation": 'Significant' if p_value < 0.05 else 'not Significant'
}


    # Interpretation
    if p_value < 0.05:
        print("There is a significant difference in acceptance levels between users who have a car and those who don't.")
    else:
        print("There is no significant difference in acceptance levels between users who have a car and those who don't.")
    if return_answer:
        return p_value < 0.05,new_row
```
