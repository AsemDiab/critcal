```python
import math
male_count=df[df["الجنس"]=="ذكر"].shape[0]
male_count
female_count=df[df["الجنس"]!="ذكر"].shape[0]
female_count
male_per=math.floor(male_count*10000/(male_count+female_count))/100
plt.pie([male_count,female_count],labels=["male "+str(male_per)+" %","female"+str(100-male_per)+" %"])
```

```python
import math
import matplotlib.font_manager as fm

# Path to the font that supports Arabic characters
# Replace 'arial.ttf' with the full path to a suitable font file
font_path = 'C:/Windows/Fonts/arial.ttf'  # Example path for DejaVuSans on Linux
the_column_name="السكن"
# Create a font object
arabic_font = fm.FontProperties(fname=font_path)
values=df[the_column_name].unique()
counts=[]
for value in values:
    counts.append(df[df[the_column_name]==value].shape[0])
counts
percent=[]
for count in counts:
    percent.append(math.floor(count*10000/sum(counts))/100)
    
percent[-1]=math.floor(10000-100*sum(percent[0:-1]))/100
chart_labels=[]
for i in range(len(values)):
    chart_labels.append(values[i]+" "+str(percent[i])+" %")    
plt.pie(counts,labels=chart_labels,textprops={'fontproperties': arabic_font})
plt.plot()
```
